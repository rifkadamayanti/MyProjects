### **1. Algoritma Clustering**

Clustering adalah alat yang berguna dalam ilmu data. Metode ini digunakan untuk menemukan struktur klaster dalam suatu kumpulan data yang ditandai dengan kesamaan terbesar di dalam klaster yang sama dan perbedaan terbesar antar klaster yang berbeda. Clustering hierarkis merupakan metode clustering pertama yang digunakan oleh para ahli biologi dan ilmuwan sosial, sedangkan analisis klaster kemudian berkembang menjadi cabang dari analisis multivariat statistik. Clustering juga merupakan pendekatan pembelajaran tanpa pengawasan (unsupervised learning) dalam machine learning [1].

Dari sudut pandang statistik, metode clustering umumnya dibagi menjadi dua pendekatan, yaitu pendekatan berbasis model probabilitas dan pendekatan nonparametrik. Pendekatan berbasis model probabilitas mengasumsikan bahwa titik data berasal dari campuran model probabilitas sehingga metode clustering menggunakan pendekatan likelihood campuran. Dalam pendekatan berbasis model, algoritma expectation and maximization (EM) adalah yang paling banyak digunakan. Sedangkan untuk pendekatan nonparametrik, metode clustering biasanya didasarkan pada fungsi objektif yang mengukur kemiripan atau ketidaksamaan, dan metode ini dapat dibagi menjadi metode hierarkis dan partisi, di mana metode partisi adalah yang paling banyak digunakan [1].

#### **1.1 K-Means Clustering**

K-means adalah metode numerik, tanpa pengawasan (unsupervised), non-deterministik, dan iteratif. Metode ini sederhana dan sangat cepat, sehingga dalam banyak aplikasi praktis, metode ini terbukti menjadi cara yang sangat efektif untuk menghasilkan hasil clustering yang baik. Namun, metode ini sangat cocok untuk menghasilkan klaster yang berbentuk globular (bulat). Beberapa upaya telah dilakukan oleh para peneliti untuk meningkatkan efisiensi algoritma k-means tersebut [2]. 

K-means adalah algoritma clustering khas dalam data mining yang banyak digunakan untuk mengelompokkan kumpulan data besar. Pada tahun 1967, MacQueen pertama kali mengusulkan algoritma k-means, yang merupakan salah satu algoritma pembelajaran tanpa pengawasan (non-supervised learning) yang paling sederhana dan diterapkan untuk menyelesaikan masalah clustering yang sudah dikenal. Algoritma ini adalah metode clustering partisi yang mengklasifikasikan objek data yang diberikan ke dalam k klaster berbeda melalui proses iteratif yang berkonvergensi ke nilai minimum lokal. Dengan demikian, hasil klaster yang dihasilkan bersifat kompak dan independen [3].

Algoritma ini terdiri dari dua fase terpisah. Fase pertama memilih k pusat klaster secara acak, dengan nilai k yang sudah ditentukan sebelumnya. Fase berikutnya adalah mengalokasikan setiap objek data ke pusat terdekat. Jarak Euclidean biasanya digunakan untuk menentukan jarak antara setiap objek data dengan pusat klaster. Ketika semua objek data telah dimasukkan ke dalam beberapa klaster, langkah pertama selesai dan pembagian awal telah dilakukan. Selanjutnya, rata-rata dari klaster yang terbentuk dihitung ulang. Proses iteratif ini terus berulang sampai fungsi kriteria mencapai nilai minimum [3].

Misalkan objek target adalah x, xᵢ menunjukkan rata-rata dari cluster Cᵢ, fungsi kriteria didefinisikan sebagai berikut:


$$ E = \sum_{i=1}^{k} \sum_{x \in C_i} |x - x_i|^2 $$

E adalah jumlah kuadrat kesalahan dari semua objek dalam database. Jarak fungsi kriteria menggunakan jarak Euclidean, yang digunakan untuk menentukan jarak terdekat antara setiap objek data dan pusat cluster. Jarak Euclidean antara satu vektor x = (x₁, x₂, ..., xₙ) dan vektor lain y = (y₁, y₂, ..., yₙ), jarak Euclidean d(xᵢ, yᵢ) dapat diperoleh sebagai berikut:

$$
d(x_i, y_i) = \left[ \sum_{i=1}^n (x_i - y_i)^2 \right]^{1/2}
$$

Proses algoritma k-means adalah sebagai berikut:

**Input:**

Jumlah cluster yang diinginkan, k, dan database D = $ \{d_1, d_2, \ldots, d_n\} $ yang berisi n objek data.

**Output:**

Satu set dari k cluster.

**Langkah-langkah:**

1. Pilih secara acak k objek data dari dataset D sebagai pusat cluster awal.
2. Ulangi;
3. Hitung jarak antara setiap objek data dᵢ (1 ≤ i ≤ n) dengan semua pusat cluster cⱼ (1 ≤ j ≤ k) dan tetapkan objek data dᵢ ke cluster terdekat.
4. Untuk setiap cluster j (1 ≤ j ≤ k), hitung ulang pusat cluster.
5. Ulangi sampai tidak ada perubahan pada pusat cluster.

Algoritma clustering k-means selalu konvergen ke minimum lokal. Sebelum algoritma k-means konvergen, perhitungan jarak dan pusat cluster dilakukan selama loop dieksekusi beberapa kali, di mana bilangan bulat positif t dikenal sebagai jumlah iterasi k-means. Nilai pasti dari t bervariasi tergantung pada pusat cluster awal yang dipilih.

Distribusi titik data memiliki hubungan dengan pusat cluster baru, sehingga kompleksitas waktu komputasi dari algoritma k-means adalah O(nkt), dengan n adalah jumlah semua objek data, k adalah jumlah cluster, dan t adalah jumlah iterasi algoritma. Biasanya berlaku k << n dan t << n.

**Kelemahan algoritma K-Means**

Dari analisis algoritma di atas, algoritma harus menghitung jarak dari setiap objek data ke setiap pusat cluster pada setiap iterasi. Namun, berdasarkan eksperimen, ternyata tidak selalu perlu menghitung jarak tersebut setiap kali.

Misalkan cluster C terbentuk setelah iterasi ke-j, dan objek data x ditetapkan ke cluster C. Pada beberapa iterasi berikutnya, objek data x tetap berada di cluster C. Dalam proses ini, setelah beberapa iterasi, kita menghitung jarak dari objek data x ke setiap pusat cluster dan menemukan bahwa jarak ke cluster C adalah yang terkecil.

Sehingga, selama beberapa iterasi, algoritma k-means harus menghitung jarak dari objek data x ke pusat cluster lain, yang memakan waktu eksekusi yang lama dan akhirnya mempengaruhi efisiensi proses clustering.

#### **1.2 K-Median Clustering** 

Metode k-median merupakan pengembangan dari metode k-means. Kedua metode ini menghasilkan k cluster dengan cara mengukur jarak antara titik pusat cluster dan setiap objek, kemudian setiap objek dikelompokkan ke pusat cluster terdekat [4].

Perbedaan utama antara keduanya terletak pada cara menentukan pusat cluster. Sesuai namanya, k-means menggunakan rata-rata (mean) sebagai pusat cluster, sedangkan k-median menggunakan median. Median sendiri merupakan ukuran statistik deskriptif yang lebih tahan terhadap nilai ekstrem (outlier) [4].

Oleh karena itu, penggunaan metode k-median dapat mengurangi kesalahan dalam pengelompokan dan menghasilkan cluster yang lebih stabil [4].

**Metode Algoritma K-medians**

1. Tentukan jumlah cluster  
   Dalam metode K-medians, jumlah cluster **k** harus ditentukan terlebih dahulu dan tidak ada aturan khusus untuk menentukan jumlah cluster **k**. Hal ini karena terkadang penentuan jumlah cluster bersifat subjektif berdasarkan peneliti. Dalam studi ini, jumlah cluster **k** ditentukan menggunakan Koefisien Silhouette. Tahapan perhitungan koefisien silhouette adalah sebagai berikut:

- Hitung rata-rata jarak objek dengan objek lain dalam cluster yang sama menggunakan persamaan:

$$
a(i) = \frac{1}{|A| - 1} \sum_{j \in A, j \neq i} d(i, j)
$$

Keterangan:  
- $a(i)$ : rata-rata jarak antar komponen dalam grup  
- $i$ : sebuah objek dalam cluster $A$ 
- $j$ : objek lain dalam cluster $A$ 
- $d(i, j)$ : jarak antara objek $i$ dan $j$  

- Hitung rata-rata jarak objek dengan semua objek di cluster lain, kemudian ambil nilai minimum dengan persamaan:

$$
d(i, C) = \frac{1}{|C|} \sum_{j \in C} d(i, j)
$$

Keterangan:  
- *d(i, C)* : rata-rata jarak antara objek *i* dengan semua objek di cluster lain *C*, dengan *A ≠ C*

- Hitung nilai koefisien silhouette dengan persamaan:

$$
s(i) = \frac{b(i) - a(i)}{\max \{a(i), b(i)\}}
$$

3. Menentukan Jarak Setiap Unit Observasi ke Setiap Centroid

Dalam metode k-medians, pengukuran jarak digunakan untuk mengelompokkan observasi berdasarkan centroid terdekat. Metode pengukuran jarak yang dipakai adalah **jarak Manhattan**. Jarak Manhattan adalah pengukuran jarak berdasarkan sistem grid di mana titik-titik yang diamati ditempatkan. Konsepnya adalah bahwa untuk berpindah dari titik awal ke titik tujuan, hanya empat arah yang dapat dipilih: atas, bawah, kiri, atau kanan. Setiap langkah akan menggerakkan titik awal satu satuan ke arah yang dipilih.

Jarak Manhattan dihitung berdasarkan banyaknya langkah (keputusan arah) yang diperlukan untuk mencapai titik tujuan dari titik awal.

Rumus jarak Manhattan dinyatakan sebagai berikut:

$$
d(x_{ij}, c_{ij}) = \sum_{j=1}^{p} |x_{ij} - c_{ij}|
\quad ; \quad i = 1, 2, \ldots, k
$$

4. Menentukan Kebaikan Metode Clustering dengan Standar Deviasi

Untuk mengetahui metode mana yang menghasilkan hasil terbaik, kita dapat menggunakan standar deviasi dalam cluster ($S_w$) dan standar deviasi antar cluster ($S_B$). Rasio variansi dinyatakan sebagai berikut:

$$
\text{varians} = \frac{S_w}{S_B} \tag{9}
$$

Rata-rata standar deviasi dalam cluster ($S_w$) dihitung dengan rumus:

$$
S_w = \frac{1}{K} \sum_{k=1}^{K} S_k \tag{10}
$$

Keterangan:  
- $K$ = jumlah cluster yang terbentuk  
- $S_k$ = standar deviasi pada cluster ke-\(k\)

Standar deviasi antar cluster (\(S_B\)) dihitung dengan rumus:

$$
S_B = \left[ \frac{1}{K-1} \sum_{k=1}^{K} \left( \bar{X}_k - \bar{X} \right)^2 \right]^{\frac{1}{2}} \tag{11}
$$

Keterangan:  
- $\bar{X}_k$ = rata-rata cluster ke-\(k\)  
- $\bar{X}$ = rata-rata keseluruhan cluster

Metode k-medians memiliki beberapa kelebihan yang membuatnya lebih unggul dalam kondisi tertentu dibandingkan metode clustering lainnya. Salah satu kelebihannya adalah k-medians lebih tahan terhadap outlier karena menggunakan median sebagai pusat cluster, sehingga nilai ekstrem tidak akan mempengaruhi hasil cluster secara signifikan seperti pada metode k-means yang menggunakan mean. Selain itu, k-medians sangat cocok digunakan pada data dengan skala ordinal atau data yang mengandung noise, karena median lebih representatif untuk data yang tidak terdistribusi secara normal. Metode ini juga umumnya menggunakan jarak Manhattan, yang sesuai untuk data berbasis grid atau data dengan dimensi yang berbeda-beda.

Namun, k-medians juga memiliki beberapa kekurangan. Perhitungan median yang harus dilakukan secara berulang membuat metode ini cenderung lebih kompleks dan memakan waktu, terutama saat diterapkan pada dataset yang sangat besar. Selain itu, pada data dengan dimensi yang sangat tinggi, penggunaan jarak Manhattan bisa kurang efektif dibandingkan jarak Euclidean. Sama seperti k-means, k-medians juga memerlukan jumlah cluster yang harus ditentukan sejak awal, yang seringkali bersifat subjektif dan sulit untuk dipastikan secara tepat. Hal ini bisa menjadi tantangan tersendiri dalam penerapan metode ini.


#### **1.3 Gaussian Mixture Model (GMM)**

Distribusi Gaussian, yang juga disebut distribusi Normal, adalah distribusi probabilitas kontinu, yang dinyatakan sebagai berikut:

$$
N(X|\mu, \Sigma) = \frac{1}{(2\pi)^{\frac{D}{2}} \sqrt{|\Sigma|}} \exp \left( -\frac{1}{2} (X-\mu)^T \Sigma^{-1} (X-\mu) \right)
\tag{3}
$$

di mana:  
- $\mu$ adalah vektor rata-rata berdimensi $D$,  
- $\Sigma$ adalah matriks kovarians $D \times D$ yang menggambarkan bentuk distribusi Gaussian,  
- $|\Sigma|$ adalah determinan dari matriks $\Sigma$.

Distribusi Gaussian bersifat simetris terhadap rata-rata dan dideskripsikan oleh rata-rata dan standar deviasi. Namun, sifat unimodal dari distribusi Gaussian tunggal tidak mampu memodelkan daerah kepadatan ganda (multimodal) pada data yang kompleks seperti pada situasi praktis. Oleh karena itu, distribusi Gaussian tunggal tidak dapat menangkap perilaku penggunaan sumber daya yang heterogen pada beban kerja Cloud. Untuk itu, distribusi yang kompleks dan multimodal dapat dimodelkan secara tepat menggunakan campuran beberapa distribusi Gaussian [5].

GMM adalah teknik clustering tanpa pengawasan yang membentuk cluster berbentuk elips berdasarkan estimasi densitas probabilitas menggunakan algoritma Expectation-Maximization (EM). Setiap cluster dimodelkan sebagai distribusi Gaussian. Berbeda dengan K-Means yang hanya menggunakan rata-rata, GMM menggunakan rata-rata dan matriks kovarians sehingga mampu memberikan ukuran kecocokan yang lebih baik terhadap jumlah cluster [5].

GMM direpresentasikan sebagai kombinasi linear dari distribusi probabilitas Gaussian dasar dan dinyatakan sebagai:

$$
p(X) = \sum_{k=1}^K \pi_k N(X|\mu_k, \Sigma_k)
\tag{4}
$$

di mana:  
- $K$ adalah jumlah komponen dalam model campuran,  
- $\pi_k$ adalah koefisien pencampur (mixing coefficient) yang memberikan estimasi densitas dari setiap komponen Gaussian,  
- $N(X|\mu_k, \Sigma_k)$ adalah distribusi Gaussian dari komponen ke-$k$ dengan rata-rata $\mu_k$ dan kovarians $\Sigma_k$.

**1.3.1 Maximum Likelihood untuk Gaussian Mixture Model**

Diberikan sekumpulan $N$ observasi independen dan identik $\{x_1, x_2, \ldots, x_N\}$, fungsi log-likelihood untuk campuran Gaussian adalah:

$$
\ln p(X|\pi, \mu, \Sigma) = \sum_{n=1}^N \ln \left( \sum_{k=1}^K \pi_k N(x_n|\mu_k, \Sigma_k) \right)
\tag{5}
$$

Penggunaan log-likelihood menghindari masalah numerik berupa underflow yang dapat terjadi akibat perkalian banyak probabilitas kecil【20】.

**1.3.2 Expectation-Maximization untuk Gaussian Mixture Model**

Algoritma EM digunakan untuk mendapatkan estimasi maksimum likelihood dari parameter GMM, yaitu vektor rata-rata $\mu$, matriks kovarians $\Sigma$, dan koefisien pencampur $\pi$ sesuai persamaa. Salah satu cara untuk menghindari algoritma EM terjebak pada maksimum lokal adalah dengan melakukan inisialisasi acak beberapa kali (multiple random initializations) karena fungsi log-likelihood dapat memiliki banyak maksimum lokal.

**1.3.3 Optimal Number of Components**

GMM adalah model generatif yang memberikan distribusi probabilitas untuk sebuah dataset.  
Jumlah komponen yang optimal sangat penting agar menghindari **overfitting** maupun **underfitting**. Jumlah ini dapat ditentukan dengan mengevaluasi likelihood model menggunakan **cross-validation** atau kriteria analitik.  

Dua metode analitik yang umum digunakan untuk mengestimasi goodness-of-fit model statistik relatif terhadap model lain pada dataset tertentu adalah **Akaike Information Criterion (AIC)** dan **Bayesian Information Criterion (BIC)**.  

Kedua metode ini memberikan ukuran kuantitatif mengenai seberapa general sebuah model, terutama dalam hal akurasi mewakili data masa depan menggunakan proses yang menghasilkan data saat ini. Baik AIC maupun BIC memberikan penalti terhadap overfitting dan underfitting, dengan nilai penalti BIC yang lebih besar dibandingkan AIC.

**1.3.4 Kelebihan dan Kekurangan Clustering dengan Gaussian Mixture Model (GMM)**

Metode Gaussian Mixture Model (GMM) memiliki beberapa kelebihan utama. Pertama, GMM mampu memodelkan cluster yang memiliki bentuk elips dan ukuran yang berbeda-beda, sehingga lebih fleksibel dibandingkan metode seperti K-Means yang hanya mampu membentuk cluster berbentuk bulat dengan ukuran seragam. Kedua, GMM memberikan pendekatan probabilistik yang memungkinkan setiap titik data memiliki probabilitas keanggotaan ke beberapa cluster sekaligus, yang berguna dalam situasi data tumpang tindih. Selain itu, GMM juga menggunakan informasi dari matriks kovarians yang menangkap variansi dan korelasi antar fitur, sehingga hasil clustering dapat lebih akurat merepresentasikan struktur data.

Namun, GMM juga memiliki beberapa kekurangan. Salah satunya adalah proses pelatihan yang lebih kompleks dan memakan waktu dibandingkan K-Means, terutama karena algoritma Expectation-Maximization (EM) yang digunakan untuk estimasi parameter dapat memerlukan banyak iterasi dan inisialisasi ulang untuk menghindari jebakan pada solusi lokal. Selain itu, GMM juga sensitif terhadap jumlah komponen yang dipilih; jika jumlah cluster tidak tepat, hasil clustering bisa kurang optimal. Terakhir, pada data dengan dimensi yang sangat tinggi, estimasi matriks kovarians bisa menjadi tidak stabil dan mempengaruhi performa model.

#### **1.4 Fuzzy C-Means (FCM)**

Fuzzy C-Means (FCM) adalah metode clustering berbasis fuzzy yang memungkinkan setiap data memiliki derajat keanggotaan terhadap beberapa cluster secara bersamaan, berbeda dengan metode clustering keras seperti K-Means yang menetapkan setiap data hanya ke satu cluster saja. Dalam FCM, setiap titik data diberikan nilai keanggotaan yang menunjukkan seberapa kuat keterkaitannya dengan masing-masing cluster, sehingga cluster yang terbentuk bersifat tumpang tindih (overlapping). Metode ini menggunakan fungsi objektif yang diminimalkan secara iteratif, dengan memperbarui pusat cluster dan nilai keanggotaan hingga konvergen. FCM banyak digunakan pada data yang memiliki batas cluster yang tidak tegas atau ambigu, sehingga memberikan fleksibilitas yang lebih tinggi dalam pemodelan struktur data.



Algoritma Fuzzy C-Means (FCM) meminimalkan fungsi objektif yang dinyatakan pada Persamaan (1):

$$
J_{FCM}(X; U, V) = \sum_{i=1}^c \sum_{j=1}^n u_{ij}^m D_{ij}^2
\tag{1}
$$

Fungsi ini berbeda dengan K-Means klasik karena menggunakan **weighted squared errors** (kesalahan kuadrat berbobot), bukan hanya kesalahan kuadrat biasa. Dalam fungsi objektif tersebut, $U$ adalah matriks partisi fuzzy yang dihitung dari dataset $X$:

$$
U = [u_{ij}] \in M_{FCM}
\tag{2}
$$

Partisi fuzzy $U$ memiliki dimensi $c \times n$, di mana $c$ adalah jumlah cluster dan $n$ adalah jumlah data. Elemen $u_{ij}$ adalah nilai keanggotaan objek ke-$j$ pada cluster ke-$i$. Kolom ke-$j$ pada matriks $U$ berisi nilai keanggotaan dari semua objek ke cluster ke-$j$.

$V$ adalah vektor prototipe atau centroid cluster:

$$
V = [v_1, v_2, ..., v_c], \quad v_i \in \mathbb{R}^p
\tag{3}
$$

Sedangkan $D_{ij}^2$ adalah jarak antara vektor fitur objek ke-$j$ dengan centroid cluster ke-$i$, yang dihitung sebagai norma jarak kuadrat dari inner-product:

$$
D_{ij}^2 = \| x_j - v_i \|_A^2 = (x_j - v_i)^T A (x_j - v_i)
\tag{4}
$$

Dalam persamaan (4), $A$ adalah matriks norma yang bersifat positif dan simetris. Inner-product dengan $A$ mengukur jarak antara titik data dan prototipe cluster. Jika $A$ adalah matriks identitas $I$, maka $D_{ij}^2$ menjadi norma Euclidean kuadrat.

Parameter $m$ dalam persamaan (1) disebut **fuzzifier** atau eksponen pembobotan, dengan nilai nyata lebih besar dari 1 ($m \in [1, \infty)$). Ketika $m$ mendekati 1, clustering cenderung menjadi keras (crisp), sedangkan ketika $m$ menuju tak hingga, clustering menjadi sangat fuzzy. Nilai fuzzifier yang umum digunakan dalam aplikasi adalah 2.

Fungsi objektif diminimalkan dengan beberapa kendala berikut:

- Nilai keanggotaan $u_{ij}$ berada di antara 0 dan 1:

$$
u_{ij} \in [0,1]; \quad 1 \leq i \leq c, \quad 1 \leq j \leq n
\tag{5}
$$

- Jumlah nilai keanggotaan tiap objek terhadap semua cluster adalah 1:

$$
\sum_{i=1}^c u_{ij} = 1; \quad 1 \leq j \leq n
\tag{6}
$$

- Jumlah nilai keanggotaan tiap cluster lebih besar dari 0 dan kurang dari jumlah total objek:

$$
0 < \sum_{j=1}^n u_{ij} < n; \quad 1 \leq i \leq c
\tag{7}
$$

FCM adalah proses iteratif yang berhenti ketika jumlah iterasi mencapai maksimum, atau ketika perbedaan antara dua nilai berturut-turut dari fungsi objektif kurang dari nilai konvergensi yang telah ditentukan ($\varepsilon$).  

Langkah-langkah dalam algoritma FCM adalah sebagai berikut:

1. Inisialisasi matriks keanggotaan $U^{(0)}$ secara acak.

2. Hitung vektor prototipe (centroid) dengan rumus:

$$
v_i = \frac{\sum_{j=1}^n u_{ij}^m x_j}{\sum_{j=1}^n u_{ij}^m}, \quad 1 \leq i \leq c
\tag{8}
$$

3. Hitung nilai keanggotaan baru menggunakan rumus:

$$
u_{ij} = \frac{1}{\sum_{k=1}^c \left( \frac{D_{ij}^A}{D_{kj}^A} \right)^{\frac{2}{m-1}}}, \quad 1 \leq i \leq c, \quad 1 \leq j \leq n
\tag{9}
$$

4. Bandingkan matriks keanggotaan $U^{(t+1)}$ dengan $U^{(t)}$, di mana $t$ adalah nomor iterasi.

5. Jika $\| U^{(t+1)} - U^{(t)} \| < \varepsilon$, maka proses berhenti; jika tidak, kembali ke langkah 2. [6]

Metode Fuzzy C-Means (FCM) memiliki kelebihan utama yaitu kemampuannya untuk memberikan derajat keanggotaan data pada setiap cluster secara bersamaan, sehingga sangat cocok untuk data dengan batas cluster yang tidak jelas atau tumpang tindih. Pendekatan fuzzy ini memungkinkan pemodelan yang lebih fleksibel dan realistis dibandingkan metode clustering keras seperti K-Means. Selain itu, FCM juga relatif mudah diimplementasikan dan dapat disesuaikan dengan berbagai jenis data melalui penggunaan matriks norma dan parameter fuzzifier.

Namun, FCM juga memiliki beberapa kekurangan. Algoritma ini sensitif terhadap nilai inisialisasi awal matriks keanggotaan, sehingga hasil clustering dapat berbeda jika inisialisasi berbeda. Selain itu, FCM cenderung terjebak pada solusi lokal karena optimisasi fungsi objektifnya bersifat non-konveks. Proses iterasi FCM juga dapat menjadi lambat terutama pada dataset besar dan berdimensi tinggi. Terakhir, FCM mengasumsikan bahwa jumlah cluster diketahui sebelumnya, yang seringkali sulit ditentukan secara tepat dalam praktik.

## Feature Selection 

### **2. Teknik Feature Selection: Variance Threshold**

Variance Threshold adalah salah satu metode feature selection yang sederhana dan efektif untuk mengurangi jumlah fitur dalam sebuah dataset. Metode ini bekerja dengan menghapus fitur-fitur yang memiliki varians (variance) di bawah nilai ambang tertentu (threshold). Fitur dengan varians rendah dianggap kurang informatif karena cenderung memiliki nilai yang hampir sama di seluruh sampel dan tidak memberikan kontribusi signifikan dalam proses pembelajaran model. Dengan menggunakan Variance Threshold, model dapat menjadi lebih sederhana, lebih cepat dilatih, dan berpotensi meningkatkan performa dengan menghilangkan fitur-fitur yang tidak relevan atau redundan.

Varians untuk sebuah fitur $ X $ dengan nilai $ x_1, x_2, \ldots, x_n $ dihitung sebagai:
$$
\text{Var}(X) = \frac{1}{n} \sum_{i=1}^n (x_i - \bar{x})^2
$$

di mana $ \bar{x} $ adalah rata-rata dari nilai fitur tersebut.

Fitur akan dipilih jika memenuhi:

$$
\text{Var}(X) \geq \theta
$$

dengan $ \theta $ adalah nilai ambang (threshold) varians yang ditentukan.

**Algoritma / Langkah-langkah**

1. Hitung varians setiap fitur pada dataset.  
2. Tentukan nilai threshold varians $ \theta $ sebagai batas minimum varians fitur yang diinginkan.  
3. Pilih fitur-fitur yang memiliki varians lebih besar atau sama dengan $ \theta $.  
4. Buang fitur yang memiliki varians lebih kecil dari $ \theta $.  
5. Gunakan subset fitur yang terpilih untuk pemodelan lebih lanjut.

**Kelebihan**

- **Sederhana dan cepat**: Mudah diimplementasikan dan sangat efisien untuk dataset dengan banyak fitur.  
- **Mengurangi dimensionalitas**: Menghilangkan fitur yang kurang informatif sehingga mempercepat proses pelatihan model dan mengurangi overfitting.  
- **Tidak memerlukan label**: Variance Threshold adalah metode unsupervised sehingga dapat diterapkan tanpa informasi target.

**Kekurangan**

- **Tidak mempertimbangkan korelasi antar fitur**: Metode ini hanya melihat varians fitur secara individual tanpa mempertimbangkan hubungan antar fitur.  
- **Bersifat statis**: Tidak menyesuaikan dengan tujuan spesifik model atau tugas prediksi tertentu.  
- **Ambang batas subjektif**: Pemilihan nilai threshold $ \theta $ bisa subjektif dan mempengaruhi hasil seleksi fitur.

Metode Variance Threshold sangat cocok sebagai langkah awal dalam proses feature selection sebelum menggunakan metode yang lebih kompleks atau supervised.

### **1.1 Algoritma K-Means**

#### **1.1.1 Menentukan Jumlah k Klaster dengan Elbow & Silhoutte Method**

Dalam proses pemodelan K-Means, digunakan dua metode untuk menentukan jumlah klaster (k) yang optimal, yaitu Elbow Method dan Silhouette Score. Elbow Method mengevaluasi total inertia (jumlah kuadrat jarak antar titik dan pusat klaster) dan mencari titik "siku" pada grafik sebagai indikasi jumlah klaster terbaik. Sementara itu, Silhouette Score mengukur seberapa baik tiap titik cocok dengan klasternya dibandingkan dengan klaster lain, dengan nilai mendekati 1 menunjukkan pemisahan klaster yang baik. 

```python
# Elbow Method untuk menentukan jumlah cluster optimal pada KMeans
inertia = []
K = range(2, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=26)
    kmeans.fit(data_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K, inertia, marker='o')
plt.title('Elbow Method: Inertia vs Jumlah Cluster (KMeans)')
plt.xlabel('Jumlah Cluster (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

**Interpretasi**

Pada grafik Elbow, terlihat penurunan nilai inertia yang sangat tajam dari k=2 ke k=3, kemudian mulai melandai setelahnya. Titik "siku" atau elbow paling jelas terlihat pada k=3, yang mengindikasikan bahwa setelah nilai tersebut, penambahan jumlah klaster tidak lagi memberikan penurunan inertia yang signifikan. Oleh karena itu, berdasarkan metode Elbow, jumlah klaster optimal adalah k = 3.

```python
from sklearn.metrics import silhouette_score

silhouette_scores = []
K = range(2, 10)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=26)
    labels = kmeans.fit_predict(data_scaled)
    score = silhouette_score(data_scaled, labels)
    silhouette_scores.append(score)

plt.figure(figsize=(10, 5))
plt.plot(K, silhouette_scores, marker='o')
plt.title('Silhouette Score vs Jumlah Cluster (KMeans)')
plt.xlabel('Jumlah Cluster (k)')
plt.ylabel('Silhouette Score')
plt.show()

# Menampilkan nilai silhouette score untuk tiap k
for k, score in zip(K, silhouette_scores):
    print(f'k={k}: Silhouette Score={score:.4f}')
```

**Interpretasi**

Untuk memvalidasi hasil pada Elbow Method, digunakan metode Silhoutte. Grafik Silhouette Score menunjukkan bahwa nilai tertinggi terjadi pada k = 8 dengan skor 0.8283, diikuti oleh k = 3 dengan skor 0.8194. Artinya, klaster dengan k = 8 memberikan pemisahan antar klaster yang paling baik menurut metrik ini. Namun, skor untuk k = 3 juga sangat tinggi dan hampir sama nilainya, ini artinya klasterisasi dengan 3 klaster masih tergolong sangat baik.


Karena kedua metode cenderung merujuk pada jumlah klaster k = 3, maka akan dipilih k = 3 sebagai jumlah klaster optimal

### **1.3 Algoritma Gaussian Mixture Models**

```python
df_gmm = data_scaled.copy()
df_gmm
```

#### **1.3.1 Menentukan Jumlah k Klaster dengan Elbow & Silhoutte Method**

Dalam proses pemodelan ini, digunakan dua metode untuk menentukan jumlah klaster (k) yang optimal, yaitu Elbow Method dan Silhouette Score. Elbow Method mengevaluasi total inertia (jumlah kuadrat jarak antar titik dan pusat klaster) dan mencari titik "siku" pada grafik sebagai indikasi jumlah klaster terbaik. Sementara itu, Silhouette Score mengukur seberapa baik tiap titik cocok dengan klasternya dibandingkan dengan klaster lain, dengan nilai mendekati 1 menunjukkan pemisahan klaster yang baik. 


```python
from sklearn.mixture import GaussianMixture

# Elbow Method untuk menentukan jumlah cluster optimal pada Gaussian Mixture Model (GMM)

sse_gmm = []
K = range(2, 11)
for k in K:
    gmm = GaussianMixture(n_components=k, random_state=26)
    gmm.fit(df_gmm)
    # Gunakan negative log likelihood sebagai "inertia" untuk GMM (semakin kecil semakin baik)
    sse_gmm.append(-gmm.score(df_gmm) * len(df_gmm))

plt.figure(figsize=(8, 5))
plt.plot(K, sse_gmm, marker='o')
plt.title('Elbow Method: Negative Log Likelihood vs Jumlah Cluster (GMM)')
plt.xlabel('Jumlah Cluster (k)')
plt.ylabel('Negative Log Likelihood')
plt.grid(True)
plt.show()
```

**Interpretasi**

Grafik menunjukkan bahwa nilai Negative Log Likelihood terus menurun seiring bertambahnya jumlah klaster (k), yang berarti model semakin fit terhadap data. Namun, penurunan paling signifikan terjadi dari k = 2 ke k = 3, lalu grafik mulai melandai setelahnya. Ini menunjukkan bahwa k = 3 merupakan titik "elbow" yang wajar, karena setelah itu penambahan klaster hanya memberi peningkatan kecil terhadap model.

#### **1.4.1 Menentukan Jumlah k Klaster dengan Elbow & Silhoutte Method**

Dalam proses pemodelan ini, digunakan dua metode untuk menentukan jumlah klaster (k) yang optimal, yaitu Elbow Method dan Silhouette Score. Elbow Method mengevaluasi total inertia (jumlah kuadrat jarak antar titik dan pusat klaster) dan mencari titik "siku" pada grafik sebagai indikasi jumlah klaster terbaik. Sementara itu, Silhouette Score mengukur seberapa baik tiap titik cocok dengan klasternya dibandingkan dengan klaster lain, dengan nilai mendekati 1 menunjukkan pemisahan klaster yang baik. 


```python
from sklearn.metrics import silhouette_score
from fcmeans import FCM

silhouette_scores_fcm = []
K = range(2, 10)
for k in K:
    fcm = FCM(n_clusters=k, random_state=26)
    fcm.fit(df_fcm.values)
    labels = fcm.predict(df_fcm.values)
    score = silhouette_score(df_fcm.values, labels)
    silhouette_scores_fcm.append(score)

plt.figure(figsize=(8, 5))
plt.plot(K, silhouette_scores_fcm, marker='o')
plt.title('Silhouette Score vs Jumlah Cluster (FCM)')
plt.xlabel('Jumlah Cluster (k)')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()

for k, score in zip(K, silhouette_scores_fcm):
    print(f'k={k}: Silhouette Score={score:.4f}')
```

**Interpretasi**

Nilai Silhouette Score tertinggi terjadi pada k = 6 dengan skor 0.8260, yang menunjukkan bahwa pemisahan klaster paling baik terjadi pada jumlah klaster tersebut. Namun, nilai untuk k = 3 juga tinggi (0.8194), hanya sedikit lebih rendah dari k = 6, dan masih mengindikasikan kualitas klaster yang sangat baik.

```python
from fcmeans import FCM

# Elbow Method untuk menentukan jumlah cluster optimal pada Fuzzy C-Means (FCM)

sse_fcm = []
K = range(2, 11)
for k in K:
    fcm = FCM(n_clusters=k, random_state=26)
    fcm.fit(df_fcm.drop(columns=['klaster'], errors='ignore').values)
    # Hitung SSE (sum of squared errors) manual dari jarak ke pusat cluster
    centers = fcm.centers
    u = fcm.u
    X = df_fcm.drop(columns=['klaster'], errors='ignore').values
    # SSE: sum over all clusters and all points of (membership^m * squared distance)
    m = 2.0
    sse = np.sum((u ** m) * ((np.linalg.norm(X[:, np.newaxis] - centers, axis=2)) ** 2))
    sse_fcm.append(sse)

plt.figure(figsize=(8, 5))
plt.plot(K, sse_fcm, marker='o')
plt.title('Elbow Method: SSE vs Jumlah Cluster (FCM)')
plt.xlabel('Jumlah Cluster (k)')
plt.ylabel('SSE')
plt.grid(True)
plt.show()
```

**Interpretasi**

Grafik SSE menunjukkan penurunan tajam dari k = 2 ke k = 3, kemudian penurunannya mulai melandai. Titik "siku" paling jelas berada di k = 3, yang menjadi indikasi bahwa penambahan klaster setelah k = 3 tidak lagi menghasilkan pengurangan error yang signifikan. Maka, dari sudut pandang Elbow Method, k = 3 merupakan jumlah klaster yang optimal.

Karena kedua metode cenderung merujuk pada jumlah klaster k = 3, maka akan dipilih k = 3 sebagai jumlah klaster optimal

#### **2.4.1 Menentukan Jumlah k Klaster dengan Elbow & Silhoutte Method**

Dalam proses pemodelan ini, digunakan dua metode untuk menentukan jumlah klaster (k) yang optimal, yaitu Elbow Method dan Silhouette Score. Elbow Method mengevaluasi total inertia (jumlah kuadrat jarak antar titik dan pusat klaster) dan mencari titik "siku" pada grafik sebagai indikasi jumlah klaster terbaik. Sementara itu, Silhouette Score mengukur seberapa baik tiap titik cocok dengan klasternya dibandingkan dengan klaster lain, dengan nilai mendekati 1 menunjukkan pemisahan klaster yang baik. 


```python
from sklearn.mixture import GaussianMixture

# Elbow Method untuk menentukan jumlah cluster optimal pada Gaussian Mixture Model (GMM)

sse_gmm = []
K = range(2, 11)
for k in K:
    gmm = GaussianMixture(n_components=k, random_state=26)
    gmm.fit(df_gmm2)
    # Gunakan negative log likelihood sebagai "inertia" untuk GMM (semakin kecil semakin baik)
    sse_gmm.append(-gmm.score(df_gmm2) * len(df_gmm2))

plt.figure(figsize=(8, 5))
plt.plot(K, sse_gmm, marker='o')
plt.title('Elbow Method: Negative Log Likelihood vs Jumlah Cluster (GMM)')
plt.xlabel('Jumlah Cluster (k)')
plt.ylabel('Negative Log Likelihood')
plt.grid(True)
plt.show()
```

**Interpretasi**

Grafik menunjukkan penurunan tajam nilai Negative Log Likelihood dari k = 2 ke k = 3, lalu menurun secara bertahap hingga mulai mendatar setelah k = 6 atau 7. Penurunan yang melambat setelah titik ini menunjukkan bahwa penambahan klaster di atas k = 6 tidak memberikan peningkatan signifikan terhadap kualitas model. Maka, titik "elbow" dapat diidentifikasi pada sekitar k = 6, yang menjadi kandidat kuat jumlah klaster optimal menurut metode ini.

```python
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score

silhouette_scores_gmm = []
K = range(2, 10)
for k in K:
    gmm = GaussianMixture(n_components=k, random_state=26)
    labels = gmm.fit_predict(data_scaled2)
    score = silhouette_score(data_scaled2, labels)
    silhouette_scores_gmm.append(score)

plt.figure(figsize=(10, 5))
plt.plot(K, silhouette_scores_gmm, marker='o')
plt.title('Silhouette Score vs Jumlah Cluster (Gaussian Mixture)')
plt.xlabel('Jumlah Cluster (k)')
plt.ylabel('Silhouette Score')
plt.show()

for k, score in zip(K, silhouette_scores_gmm):
    print(f'k={k}: Silhouette Score={score:.4f}')
```

#### **2.5.1 Menentukan Jumlah k Klaster dengan Elbow & Silhoutte Method**

Dalam proses pemodelan ini, digunakan dua metode untuk menentukan jumlah klaster (k) yang optimal, yaitu Elbow Method dan Silhouette Score. Sementara itu, Silhouette Score mengukur seberapa baik tiap titik cocok dengan klasternya dibandingkan dengan klaster lain, dengan nilai mendekati 1 menunjukkan pemisahan klaster yang baik. 


```python
from sklearn.metrics import silhouette_score
from fcmeans import FCM

silhouette_scores_fcm = []
K = range(2, 10)
for k in K:
    fcm = FCM(n_clusters=k, random_state=26)
    fcm.fit(df_fcm2.values)
    labels = fcm.predict(df_fcm2.values)
    score = silhouette_score(df_fcm2.values, labels)
    silhouette_scores_fcm.append(score)

plt.figure(figsize=(8, 5))
plt.plot(K, silhouette_scores_fcm, marker='o')
plt.title('Silhouette Score vs Jumlah Cluster (FCM)')
plt.xlabel('Jumlah Cluster (k)')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()

for k, score in zip(K, silhouette_scores_fcm):
    print(f'k={k}: Silhouette Score={score:.4f}')
```

Nilai Silhouette Score tertinggi terjadi pada k = 6 dengan skor 0.8527, yang menunjukkan bahwa pemisahan klaster paling baik terjadi pada jumlah klaster tersebut. Namun, nilai untuk k = 3 juga tinggi (0.8373), hanya sedikit lebih rendah dari k = 6, dan masih mengindikasikan kualitas klaster yang sangat baik.

```python
from fcmeans import FCM

# Elbow Method untuk menentukan jumlah cluster optimal pada Fuzzy C-Means (FCM)

sse_fcm = []
K = range(2, 11)
for k in K:
    fcm = FCM(n_clusters=k, random_state=26)
    fcm.fit(df_fcm2.drop(columns=['klaster'], errors='ignore').values)
    # Hitung SSE (sum of squared errors) manual dari jarak ke pusat cluster
    centers = fcm.centers
    u = fcm.u
    X = df_fcm2.drop(columns=['klaster'], errors='ignore').values
    # SSE: sum over all clusters and all points of (membership^m * squared distance)
    m = 2.0
    sse = np.sum((u ** m) * ((np.linalg.norm(X[:, np.newaxis] - centers, axis=2)) ** 2))
    sse_fcm.append(sse)

plt.figure(figsize=(8, 5))
plt.plot(K, sse_fcm, marker='o')
plt.title('Elbow Method: SSE vs Jumlah Cluster (FCM)')
plt.xlabel('Jumlah Cluster (k)')
plt.ylabel('SSE')
plt.grid(True)
plt.show()
```

**Interpretasi**

Grafik menunjukkan penurunan SSE (Sum of Squared Error) yang sangat tajam dari k = 2 ke k = 3, kemudian terus menurun namun mulai melandai setelah k = 6. Titik elbow paling mencolok terlihat pada k = 3.

### **1. Fungsi untuk metrik evaluasi clustering (DBI, CHS, dan Silhoutte)**

```python
# Fungsi untuk metrik evaluasi clustering (DBI, CHS, dan Silhoutte)
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

def evaluate_clustering(X, cluster_labels):
    """
    Fungsi untuk menghitung beberapa metrik evaluasi clustering:
    - Silhouette Score
    - Davies-Bouldin Index (DBI)
    - Calinski-Harabasz Index (CH)

    Parameters:
    - X: Feature
    - cluster_labels: Label hasil clustering (hasil prediksi dari KMeans atau algoritma lainnya)

    Returns:
    - dict: Metrik evaluasi clustering
    """

    # Silhouette Score
    silhouette = silhouette_score(X, cluster_labels)

    # Davies-Bouldin Index (DBI)
    dbi = davies_bouldin_score(X, cluster_labels)

    # Calinski-Harabasz Index (CH)
    ch = calinski_harabasz_score(X, cluster_labels)

    # Return all metrics in a dictionary
    return {
        'Silhouette Score': silhouette,
        'Davies-Bouldin Index (DBI)': dbi,
        'Calinski-Harabasz Index (CH)': ch
    }
```

#### **Fungsi Dunn-Index**

```python
from scipy.spatial.distance import cdist

def dunn_index(X, labels):
    """
    Menghitung Dunn Index untuk hasil clustering.
    Semakin besar nilai Dunn Index, semakin baik hasil clustering.
    """
    unique_clusters = np.unique(labels)
    clusters = [X[labels == k] for k in unique_clusters if k != -1]  # abaikan noise jika ada label -1

    # Hitung jarak minimum antar cluster (inter-cluster)
    min_intercluster = np.inf
    for i in range(len(clusters)):
        for j in range(i + 1, len(clusters)):
            dist = cdist(clusters[i], clusters[j])
            min_dist = np.min(dist)
            if min_dist < min_intercluster:
                min_intercluster = min_dist

    # Hitung diameter maksimum dalam cluster (intra-cluster)
    max_intracluster = 0
    for cluster in clusters:
        if len(cluster) > 1:
            dist = cdist(cluster, cluster)
            max_dist = np.max(dist)
            if max_dist > max_intracluster:
                max_intracluster = max_dist

    # Hindari pembagian dengan nol
    if max_intracluster == 0:
        return np.nan

    return min_intercluster / max_intracluster
```

### **Identifikasi Apakah Ada Cluster Berbeda yang Memiliki Provinsi Sama**

```python
# Print sorted province names for each cluster
print("Cluster 1:", sorted(provinsi_cluster1))
print("Cluster 2:", sorted(provinsi_cluster2))
print("Cluster 3:", sorted(provinsi_cluster3))
print("Cluster 4:", sorted(provinsi_cluster4))
print("Cluster 5:", sorted(provinsi_cluster5))
print("Cluster 6:", sorted(provinsi_cluster6))

# Check for overlaps between clusters
clusters = [
    provinsi_cluster1,
    provinsi_cluster2,
    provinsi_cluster3,
    provinsi_cluster4,
    provinsi_cluster5,
    provinsi_cluster6
]

overlap_found = False
for i in range(len(clusters)):
    for j in range(i+1, len(clusters)):
        overlap = clusters[i].intersection(clusters[j])
        if overlap:
            print(f"Overlap between Cluster {i+1} and Cluster {j+1}: {sorted(overlap)}")
            overlap_found = True

if not overlap_found:
    print("Tidak ada provinsi yang sama antar cluster (tidak ada overlap).")
```

- **Jawa Timur (Cluster 6)**: Juga memiliki **populasi besar dan angka kasus serta kematian yang tinggi**, terutama dalam **kematian baru**. Termasuk provinsi dengan **beban pandemi signifikan** yang perlu penguatan mitigasi dan pelayanan kesehatan.

- **Banten dan DIY (Cluster 5)**: Menunjukkan nilai **menengah dalam sebagian besar indikator**, namun **kematian per juta penduduk cukup tinggi**, yang bisa mengindikasikan tekanan pada sistem kesehatan atau populasi rentan. Perlu perhatian dalam aspek perlindungan kelompok rentan dan akses layanan kesehatan.

- **Bali, Bengkulu, Gorontalo, Kalimantan Selatan, Kepulauan Riau, Lampung, NTB, NTT, Sulawesi Selatan, Sulawesi Utara, Sumatera Barat, Sumatera Utara (Cluster 1)**: Termasuk provinsi dengan **kasus dan kematian baru yang rendah**, serta **tingkat kesembuhan sedang**. Kemungkinan wilayah dengan penyebaran yang lebih terkendali atau aktivitas sosial yang lebih terbatas.

- **Aceh, Jambi, Kalimantan Barat, Kalimantan Tengah, Kalimantan Timur, Kalimantan Utara, Babel, Maluku, Maluku Utara, Papua, Papua Barat, Riau, Sulawesi Barat, Sulawesi Tengah, Sulawesi Tenggara, Sumatera Selatan (Cluster 2)**: Menunjukkan **nilai-nilai terendah hampir di semua indikator**. Kemungkinan merupakan wilayah dengan **sebaran penduduk lebih merata atau terbatas**, aktivitas sosial-ekonomi lebih rendah, atau pelaporan kasus yang lebih minim.

**Rekomendasi**

- **Wilayah seperti DKI Jakarta, Jawa Barat, Jawa Tengah, dan Jawa Timur** memerlukan **fokus utama dalam penguatan kapasitas sistem kesehatan**, mengingat tingginya kasus dan kematian yang terjadi. Strategi seperti peningkatan kapasitas rumah sakit, percepatan vaksinasi booster, dan penegakan protokol kesehatan sangat penting di wilayah dengan mobilitas tinggi dan kepadatan penduduk besar. Selain itu, pendekatan berbasis komunitas untuk menjangkau populasi yang sulit diakses juga harus diperkuat.

- **Wilayah dengan tingkat pelaporan rendah seperti provinsi dalam Cluster 1 dan 2** perlu ditingkatkan **sistem surveilans dan kapasitas pelaporan data**. Hal ini penting agar pemerintah pusat dapat memperoleh gambaran yang lebih akurat mengenai kondisi pandemi secara nasional dan merancang intervensi yang tepat sasaran. Diperlukan pelatihan tenaga kesehatan daerah dan peningkatan infrastruktur digital untuk pelaporan yang lebih akurat dan real-time.

- Untuk provinsi seperti **Banten dan DIY**, perlu dilakukan **evaluasi terhadap kelompok rentan**, termasuk lansia dan individu dengan komorbid, yang kemungkinan berkontribusi pada tingginya kematian per juta penduduk. Intervensi yang lebih terarah seperti pelayanan kesehatan berbasis rumah dan distribusi alat kesehatan penting (seperti oksigen) bisa sangat membantu mengurangi angka kematian pada kelompok ini.

- Dengan segmentasi yang jelas berdasarkan hasil clustering, pemerintah dapat **menerapkan kebijakan berbasis data yang lebih spesifik dan efisien**, sesuai dengan karakteristik masing-masing wilayah, guna mengoptimalkan pengendalian pandemi secara nasional.
